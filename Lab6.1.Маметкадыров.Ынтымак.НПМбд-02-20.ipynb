```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = Image.open(r'C:\Users\\PycharmProjects\homework06\Bliss.bmp')
img
```

```python
x = np.array(img, dtype=np.float32)
x.dtype
```

```python
x.shape
```

```python
y = Image.fromarray(np.asarray(x.mean(axis=2), dtype=np.uint8))
y
```

```python
U, S, V = np.linalg.svd(y, full_matrices=False)
```

```python
U.shape
```

```python
B = U.dot(np.diag(S).dot(V))
B.shape
```

```python
C = Image.fromarray(np.asarray(B, dtype=np.uint8))
C
```

```python
B.shape
```

```python
S = np.diag(S)
```

```python
r = [10, 20, 30, 40, 60, 80]
for i in r:
    new_y = U[:,:i] @ S[:i,:i] @ V[:i,:]
    new_y[new_y < 0] = 0
    new_y[new_y > 255] = 255
    Image.fromarray(np.asarray(new_y, dtype=np.uint8)).save(f'result_{i}.png')
```
