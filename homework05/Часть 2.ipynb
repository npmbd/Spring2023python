## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
```

```python
xx = np.linspace(-1, 1, 250)
```

```python
yy = xx**2*np.sin(1/xx**2) + xx
```

```python
plt.plot(xx, yy, label='func')
plt.legend()
plt.xlabel('Ось x', family='verdana')
plt.ylabel('Ось y', family='verdana')
```

```python
def f1(x):
    return 1 / (1 + x**2)

def f2(x):
    return 1 / x**2

x = np.linspace(5, 25, 1000)
y1 = f1(x)
y2 = f2(x)
relative_diff = np.abs(1 - y1/y2)

plt.semilogy(x, relative_diff)
plt.xlabel('x')
plt.ylabel('Относительная разница')
plt.title('Относительная разница функций')
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = special.factorial(x)
plt.plot(x, y, label='Factorial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
x = np.linspace(1, 10, 100)
y = special.exp1(x)
plt.plot(x, y, label='Exponential integral E1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
fig, ax = plt.subplots()
xx = np.arange(-6*np.pi, 6*np.pi, np.pi/2)
x = np.linspace(-6*np.pi, 6*np.pi, 200)
y = np.sin(x)/x
ax.plot(x, y)
ax.set_xticks(xx)

labels = []
num = -12
for i in range(len(xx)):
    if num%2 == 0 and num != -2 and num != 2 and num/2 != 0:
        labels.append(f"{int(num/2)}\u03C0")
    elif num/2 == 0:
        labels.append('0')
    elif num/2 == -1.0:
        labels.append("-\u03C0")
    elif num/2 == 1.0:
        labels.append("\u03C0")
    elif int(num) == -1:
        labels.append("-\u03C0/2")
    elif int(num) == 1:
        labels.append("\u03C0/2")
    else:
        labels.append(f"{num}\u03C0/2")
    num += 1
ax.set_xticklabels(labels)
ax.grid()
fig.set_figwidth(14)
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x, y = np.meshgrid(np.linspace(-np.pi, np.pi, 200), np.linspace(-np.pi, np.pi, 200))
z = np.sin(x)*np.sin(y)
fig, axs = plt.subplots()
axs.pcolormesh(x, y, z)
axs.set_title('sin(x)sin(y)')
```

```python
x, y = np.meshgrid(np.linspace(-np.pi, np.pi, 200), np.linspace(-np.pi, np.pi, 200))
z = np.exp(-x**2 - y**2)
fig, axs = plt.subplots()
axs.pcolormesh(x, y, z)
axs.set_title('exp(-x^2 - y^2)')
```
