```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 15, 100)
y = np.sin(x/5)*np.e**(x/10) + 5*np.e**(-x/2)
plt.plot(x, y)
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

x = np.linspace(1, 15, 100)
y = f(x)

n = 15
A = np.zeros((n, n))
b = np.zeros(n)

for i in range(n):
    for j in range(n):
        A[i][j] = (i+1)**j

    b[i] = f(i+1)

coefficients = np.linalg.solve(A, b)

plt.plot(x, y, label='Исходная функция')
plt.plot(np.arange(1, 16), f(np.arange(1, 16)), 'o', label='Точки')
plt.plot(x, np.polyval(coefficients[::-1], x), label='Полиномиальная аппроксимация')
plt.legend()
plt.show()
```

При n = 15 полином хорошо приближает исходную функцию. Рассмотрели случаю, когда n = 5 и n = 15. С ростом числа точек полином приближает исходную функцию точнее. Как видно из графика выше, при добавлении шума точность приближения снижается.

```python
def f(x):
    return np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)

N = 100
x = np.linspace(1, 15, N)
y = f(x)

d = 0.1
noise = np.random.normal(0, d, N)
y_noisy = y + noise

fig, ax = plt.subplots(figsize=(8, 6))

x_true = np.linspace(1, 15, 1000)
y_true = f(x_true)
ax.plot(x_true, y_true, 'r-', label='True function')

ax.plot(x, y_noisy, 'bo', label='Data points')

A = np.vander(x, 4) # матрица коэффициентов
p = np.linalg.lstsq(A, y_noisy, rcond=None)[0] # решение системы уравнений
y_poly = np.polyval(p, x_true)
ax.plot(x_true, y_poly, 'g--', label='Polynomial fit')

ax.legend(loc='upper left')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
